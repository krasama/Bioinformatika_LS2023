```python

r_df=df.iloc[:,[0,1]]
r_df
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

smiles_colum = 'smiles'

radius = 2
nBits = 2048

r_df['fingerprint'] = r_df['smiles'].apply(lambda x: AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(x),radius, nBits))

#fps = []
#for smi in r_df[smiles_colum]:
#    mol = Chem.MolFromSmiles(smi)
#    fp = AllChem.GetMorganFingerprintAsBitVect(mol,radius,nBits)
#    fps.append(fp)
#r_df['fingerprints'] = fps
r_df
```

```python
from rdkit import DataStructs as ds

reference_smile='C1=CC(=C(C=C1CC(C(=O)O)N)O)O'

reference_fp = AllChem.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(reference_smile),radius,nBits)
r_df['similarity']=r_df['fingerprint'].apply(lambda fp: ds.TanimotoSimilarity(fp, reference_fp))
r_df
```

```python
top_3 = r_df.sort_values(by='similarity', ascending=False).head(3)
top_3
```
