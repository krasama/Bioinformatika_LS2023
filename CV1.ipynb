```python
from rdkit import DataStructs as ds

reference_smile="C1=CC(=C(C=C1CC(C(=O)O)N)O)O"

reference_fp = Chem.RDKFingerprint(Chem.MolFromSmiles(reference_smile))
r_df['similarity']=r_df['fingerprint'].apply(lambda fp: ds.TanimotoSimilarity(fp, reference_fp))
r_df
```

```python
top_3 = r_df.sort_values(by='similarity', ascending=False).head(3)
top_3
```

V dalších dvou top 3 případech je podobnost z 66% protože tyto dvě sloučeniny se liší jen v jedné vazbě, jinak je jejich obsah stejný.

```python
from scipy.spatial.distance import dice, euclidean
from rdkit import DataStructs

# funkce pro výpočet Tanimotovy míry
def tanimoto(fp1, fp2):
    return DataStructs.TanimotoSimilarity(fp1, fp2)

# funkce pro výpočet Diceovy míry
def dice_similarity(fp1, fp2):
    return 1 - dice(fp1, fp2)

# funkce pro výpočet Euklidovské vzdálenosti
def euclidean_distance(fp1, fp2):
    return euclidean(fp1, fp2)


r_df['tanimoto_similarity'] = r_df['fingerprint'].apply(lambda fp: tanimoto(fp, reference_fp))
r_df['dice_similarity'] = r_df['fingerprint'].apply(lambda fp: dice_similarity(fp, reference_fp))
r_df['euclidean_distance'] = r_df['fingerprint'].apply(lambda fp: euclidean_distance(fp, reference_fp))
r_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# seznam fingerprintů
fps = r_df['fingerprint']

# seznam similarity měřítek
similarity_measures = ['Tanimoto', 'Dice', 'Euclidean']

# vytvoření prázdného pole pro hodnoty podobnosti
similarities = np.zeros((len(similarity_measures), len(fps)))

# výpočet podobnosti mezi fingerprinty
for i, fp in enumerate(fps):
    similarities[0, i] = tanimoto(fp, reference_fp)
    similarities[1, i] = dice_similarity(fp, reference_fp)
    similarities[2, i] = euclidean_distance(fp, reference_fp)

# vykreslení grafů pro každý fingerprint zvlášť
for i, fp in enumerate(fps):
    fig, ax = plt.subplots()
    ax.bar(similarity_measures, similarities[:, i])
    ax.set_title(r_df['smiles'].iloc[i])
    ax.set_xlabel('Similarity Measure')
    ax.set_ylabel('Similarity Score')
    plt.show()
```
